```python
import re
import pickle
import time
from tkinter import Tk, Label, Entry, Button
from tkinter import ttk
from nltk.stem.porter import PorterStemmer


with open("svm_model.pkl", 'rb') as model_file: 
    loaded_model = pickle.load(model_file)
with open("tfidf_vectorizer.pkl", 'rb') as vectorizer_file:  
    cv = pickle.load(vectorizer_file)

def predict_article_category(article):
    # Process the article (cleaning, TF-IDF conversion)
    article = re.sub('[^a-zA-Z]', ' ', article)  
    article = article.lower()  
    ps = PorterStemmer()
    article = ' '.join([ps.stem(word) for word in article.split()])
    article_vector = cv.transform([article]).toarray()

    # Predict Category
    prediction = loaded_model.predict(article_vector)
    if prediction == 0:
        return "Business"
    elif prediction == 1:
        return "Entertainment"
    elif prediction == 2:
        return "Health"
    elif prediction == 3:
        return "News"
    elif prediction == 4:
        return "Politics"
    elif prediction == 5:
        return "Sport"
    else:
        return "Unknown"

def handle_click():
    article_text = article_entry.get()
    category = predict_article_category(article_text)
    
    # Animation effect
    result_label.config(text="")
    for i in range(4):
        result_label.config(text="Predicting" + "." * (i % 3 + 1))
        root.update_idletasks()
        time.sleep(0.5)
    
    result_label.config(text=f"Category: {category}")
    update_status("Prediction Complete")

# Color palette based on #FF33AF
base_color = '#FF33AF'
light_color = '#FF99D6'  # Lighter shade
dark_color = '#CC298B'   # Darker shade
bg_color = '#F0E6F2'     # Background color

# Create the main window
root = Tk()
root.title("Article Classification")

# Set window size
root.geometry("600x400")

# Set background color
root.configure(bg=bg_color)

# Create styles for ttk widgets
style = ttk.Style()
style.configure('TLabel', background=bg_color, foreground=dark_color, font=('Arial', 12, 'bold'))
style.configure('TButton', background=base_color, foreground='#000000', font=('Arial', 12, 'bold'))
style.configure('TEntry', foreground=dark_color, font=('Arial', 12))

# Create a frame for better layout management
frame = ttk.Frame(root, padding="20", style='TFrame')
frame.pack(expand=True)

# Create labels
instruction_label = ttk.Label(frame, text="Enter Article:")
instruction_label.grid(row=0, column=0, columnspan=2, pady=10)

# Create entry field for article
article_entry = ttk.Entry(frame, width=50)
article_entry.grid(row=1, column=0, columnspan=2, pady=10)

# Create a button
predict_button = ttk.Button(frame, text="Predict Category", command=handle_click)
predict_button.grid(row=2, column=0, columnspan=2, pady=20)

# Add a status bar for additional user feedback
status_bar = ttk.Label(frame, text="Ready", anchor="center", font=('Arial', 10), background=light_color, foreground=dark_color)
status_bar.grid(row=3, column=0, columnspan=2, pady=10)

# Create result label and position it below the entry field
result_label = ttk.Label(frame, text="", foreground=base_color)
result_label.grid(row=4, column=0, columnspan=2, pady=10)

def update_status(message):
    status_bar.config(text=message)
    root.update_idletasks()

# Hook into the entry widget to update the status bar
def on_entry_change(*args):
    update_status("Typing...")

article_entry.bind("<KeyRelease>", on_entry_change)

# Run the main loop
root.mainloop()
```
